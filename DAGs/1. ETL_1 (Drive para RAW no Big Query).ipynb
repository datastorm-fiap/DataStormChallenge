## 2022

```python
# Caminho para o arquivo CSV

caminho_csv = '/content/drive/MyDrive/Big Query RAW Tables/Interrupcoes/interrupcoes-energia-eletrica-2022.csv'

df_interrupcoes_2022 = pd.read_csv(caminho_csv,
            encoding='ISO-8859-1',  # Codificação do arquivo
            on_bad_lines='skip',     # Ignora linhas que não podem ser analisadas corretamente
            delimiter=';'             # Define o delimitador
        )

# Parte do texto que você está procurando, independentemente de maiúsculas/minúsculas
textos = ['meio ambiente', 'Meio Ambiente', 'MEIO AMBIENTE']

# Criar uma expressão regular de busca
# Juntar todos os termos com '|' que significa 'ou' em regex
padrao_regex = '|'.join(textos)

# Filtrar o DataFrame sem se preocupar com maiúsculas ou minúsculas
df_interrupcoes_2022 = df_interrupcoes_2022[
    df_interrupcoes_2022['DscFatoGeradorInterrupcao'].str.contains(padrao_regex, case=False, na=False)
]

# Parte do texto que você está procurando, independentemente de maiúsculas/minúsculas
textos = ['eletropaulo']

# Criar uma expressão regular de busca
# Juntar todos os termos com '|' que significa 'ou' em regex
padrao_regex = '|'.join(textos)

# Filtrar o DataFrame sem se preocupar com maiúsculas ou minúsculas
df_interrupcoes_2022 = df_interrupcoes_2022[
    df_interrupcoes_2022['NomAgenteRegulado'].str.contains(padrao_regex, case=False, na=False)
]

df_interrupcoes_2022['DscFatoGeradorInterrupcao'] = df_interrupcoes_2022['DscFatoGeradorInterrupcao'].str.upper()
df_interrupcoes_2022['DscFatoGeradorInterrupcao'] = df_interrupcoes_2022['DscFatoGeradorInterrupcao'].str.replace('INTERNA;NAO PROGRAMADA;MEIO AMBIENTE;', '', regex=False)
df_interrupcoes_2022['DscFatoGeradorInterrupcao'] = df_interrupcoes_2022['DscFatoGeradorInterrupcao'].str.replace('INTERNA - NAO PROGRAMADA - MEIO AMBIENTE - ', '', regex=False)

def remove_accents(text):
    if isinstance(text, str):
        return unidecode(text)
    return text

# Aplicar a função para remover acentos na coluna desejada
df_interrupcoes_2022['DscConjuntoUnidadeConsumidora'] = df_interrupcoes_2022['DscConjuntoUnidadeConsumidora'].apply(remove_accents)

# Deletando linhas que contêm "ANIMAIS" e "QUEIMA OU INCENDIO"
df_interrupcoes_2022 = df_interrupcoes_2022[~df_interrupcoes_2022['DscFatoGeradorInterrupcao'].str.contains('ANIMAIS|QUEIMA OU INCENDIO', case=False, na=False)]
```

# ETL Temperatura e Umidade

```python

# Definir o caminho do diretório
caminho = '/content/drive/MyDrive/Big Query RAW Tables/Temp_UR'

# Listar e filtrar os arquivos que começam com '20' e terminam com '.xlsx'
arquivos = [f for f in os.listdir(caminho) if f.startswith('temp') and f.endswith('.xlsx')]

# Inicializar uma lista para armazenar os DataFrames
temp_total = []

# Iterar sobre os arquivos e carregá-los como DataFrames
for arquivo in arquivos:
    caminho_arquivo = os.path.join(caminho, arquivo)
    df = pd.read_excel(caminho_arquivo)  # Carregar cada arquivo .xlsx como DataFrame
    temp_total.append(df)

# Verificar se a lista de DataFrames não está vazia antes de concatenar
if temp_total:
    df_temp_ur = pd.concat(temp_total, ignore_index=True)
    print("DataFrames concatenados com sucesso.")
else:
    print("Nenhum DataFrame foi carregado para concatenar.")
```

```python
# Usando melt para transformar as colunas em linhas
df_melted = df_temp_ur.melt(id_vars=['Data', 'Variavel'], var_name='SUB', value_name='Valor')

# Usando pivot para transformar o DataFrame
df_temp_ur = df_melted.pivot(index=['Data', 'SUB'], columns='Variavel', values='Valor').reset_index()

# Exibindo o resultado
df_temp_ur.sample(15)
```

# ETL Alagamentos

```python
# Transforma um excel com multiplas abas de dados (por ano) em um único dataframe consolidado
df_alagamento = pd.read_excel('/content/drive/MyDrive/Big Query RAW Tables/ALAGAMENTOS_CGE_2022_2024.xlsx', sheet_name=None)
df_alagamento = pd.concat(df_alagamento.values(), ignore_index=True)

# Lista de colunas que serão limpas
colunas_limpas = ['LOCAL', 'REF', 'CONDICAO', 'SUB']

# Remover vírgulas, pontos, asteriscos e espaços de início e fim das colunas
for column in colunas_limpas:
    df_alagamento[column] = (
        df_alagamento[column]
        .str.replace(',', '', regex=False)
        .str.replace('.', '', regex=False)
        .str.replace('*', '', regex=False)
        .str.strip()
    )

def remove_accents(text):
    if isinstance(text, str):
        return unidecode(text)
    return text

# Aplicar a função para remover acentos na coluna 'COLUNA_EXEMPLO'
df_alagamento['LOCAL'] = df_alagamento['LOCAL'].apply(remove_accents)
df_alagamento['REF'] = df_alagamento['REF'].apply(remove_accents)
df_alagamento['CONDICAO'] = df_alagamento['CONDICAO'].apply(remove_accents)

df_alagamento.tail()

# Remover linhas em branco em casos de colunas com infos indispensáveis
df_alagamento = df_alagamento.dropna(subset=['LOCAL'])
df_alagamento = df_alagamento.dropna(subset=['SUB'])
df_alagamento = df_alagamento.dropna(subset=['CONDICAO'])

#Padronização da coluna CONDICAO
df_alagamento['CONDICAO'] = df_alagamento['CONDICAO'].str.upper()
df_alagamento['CONDICAO'] = df_alagamento['CONDICAO'].str.replace('INTRANSITAVEL INTRANSITAVEL INTRANSITAVEL', 'INTRANSITAVEL', regex=False)
df_alagamento['CONDICAO'] = df_alagamento['CONDICAO'].str.replace('TRANSITAVEL (I)', 'TRANSITAVEL', regex=False)

```

## Carregar e Concatenar arquivos DBF

```python
# Diretório onde os arquivos DBF estão localizados
pasta = '/content/drive/MyDrive/Big Query RAW Tables/Ocorrencias'

# Lista para armazenar os DataFrames lidos
dataframes = []

# Loop através dos arquivos no diretório
for filename in os.listdir(pasta):
    if filename.endswith('.dbf'):  # Filtrar apenas arquivos .dbf
        # Caminho completo do arquivo
        dbf_file = os.path.join(pasta, filename)

        # Tentar ler o arquivo DBF
        try:
            # Usar a codificação 'utf-8'
            dbf = DBF(dbf_file, encoding='utf-8')
            df = pd.DataFrame(iter(dbf))
            dataframes.append(df)
            print(f"Lido: {filename}")  # Informar que o arquivo foi lido com sucesso
        except Exception as e:
            print(f"Erro ao ler {filename}: {e}")

# Concatenar todos os DataFrames em um único DataFrame, se houver DataFrames para concatenar
if dataframes:
    df_ocorrencias = pd.concat(dataframes, ignore_index=True)

    # Exibir o resultado
    df_ocorrencias.head()

else:
    print("Nenhum DataFrame foi lido.")
```
